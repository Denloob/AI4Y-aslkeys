# Study of the data

```python
import pandas as pd
df = pd.read_csv('data/dataset.csv')
display(df.info())
display(df.describe())
display(df)
print(f'{df.shape=}')
print(f'unique values of 17 disease withoud nan: {df["Symptom_17"].dropna().unique()}')
```

* Dataset has 4920 rows and 18 columns, one of them is Diseases and others are Symptoms.
* Symptoms 1-3 are allways not null.
* If disease has sympotm in column X all the previous symptoms are not null.
* Symtoms 4-17 are nullable.
* Most (>50%) of diseases have also symtopms 4-6.
* Also a litlle less than 50% of diseases also have symptom 7.
* If a disease has symptom 17, it is always muscle_pain.

```python
# convert the entire dataset to unic numeric values

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

numeric_df = df.copy()
for i, colomn in enumerate(numeric_df.columns):
    numeric_df[colomn] = le.fit_transform(df[colomn])

display(numeric_df.info())
display(numeric_df.describe())
display(numeric_df)
print(f'{numeric_df.shape=}')
```

```python
# make the model and train it based on the dataset using KNeighborsClassifier from sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X, y = numeric_df.drop('Disease', axis=1), numeric_df['Disease']
X_train, X_test, y_train, y_test = train_test_split(
 X, y, test_size=0.33, random_state=42)

# it shouldn't work like this WTF
KNN = KNeighborsClassifier(n_neighbors=1)

model = KNN.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
# check the accuracy of the model
print(f'accuracy={accuracy_score(y_test, model.predict(X_test))}')
```

```python
input_df = pd.DataFrame(columns=X_test.columns)
# itching, headache, nodal_skin_eruptions, null, null, null, ...
input_df.loc[1] = [33, 17, 35, 50, 38, 32, 26, 21, 22, 21, 18, 11, 8, 4, 3, 3, 1]
input_df
```
